The **`%`** symbol denotes a **"magic command"** in Jupyter. These are special instructions that provide functionality beyond standard Python. 

* **In this notebook:** We use `%pip install` to ensure libraries like *PyTorch*, *torchvision*, and *transformers* are installed specifically in the current environment. 
* **Who it talks to:** The **IPython interpreter**.

---

A similar symbol is **`!`**, which runs **shell commands** directly on the operating system. While `!pip install` works, **`%pip install`** is the recommended choice in Colab/Jupyter because it guarantees the package is installed in the active kernel's environment, avoiding version conflicts.

* **Who it talks to:** Your computer’s **Operating System** (Terminal/CMD/Bash).

#### **Other Popular Magic Commands:**
* `%matplotlib inline`: Displays plots directly below the code cell.
* `%time`: Measures the execution time of a single
* `%%time`: Measures the execution time of an entire cell.
* `%load_ext autoreload`: Automatically reloads modules before executing code.
* `%autoreload 2`: Reloads all modules every time before executing code.
* `%who`: Lists all variables in the current namespace.

## Task 3: Activating the GPU (Graphics Processing Unit)

Why do we need a GPU and what the heck is it?

<details>
<summary>Click here to see the answer</summary>

A GPU (Graphics Processing Unit) is a processor originally designed for rendering graphics, but it excels at deep learning because it can perform thousands of calculations in parallel. This is essential for training deep learning models efficiently, as neural networks rely heavily on matrix multiplications that GPUs handle much faster than CPUs [[3](https://www.datacamp.com/blog/cpu-vs-gpu)].

</details>

Let's activate the GPU in Colab:
1. In your Colab notebook, click on **Runtime** in the top menu.
2. Select **Change runtime type**.
3. In the pop-up window, under **Hardware accelerator**, select **GPU**.
4. Click **Save**.

Now, let's write our first bit of code to verify we are using the GPU with [PyTorch](https://pytorch.org/)!

```python
import torch

# This tells PyTorch to use the GPU if it's available, otherwise fallback to CPU
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"

print(f"You are currently using a: {device.upper()}")
```

## Task 4: Loading a PyTorch Model (ResNet)

ResNet (Residual Network) is a famous CNN (Convolutional Neural Network) model used for image classification. You can read an overview [[4](https://www.geeksforgeeks.org/deep-learning/residual-networks-resnet-deep-learning/)] or the original research paper [[5](https://arxiv.org/abs/1512.03385)].

PyTorch makes it incredibly easy to load pre-trained models so let's do it!

```python
import torchvision.models as models

# 1. Load the ResNet-18 model. "weights='DEFAULT'" loads a pre-trained version.
resnet = models.resnet18(weights='DEFAULT')
print(type(resnet))

# 2. Move the model to our device (GPU)
resnet = resnet.to(device)
resnet.eval()  # Set the model to evaluation mode (important for inference)

print("ResNet successfully loaded and moved to:", device)
```
